```python

import yfinance as yf
import pandas as pd
import numpy as np
from scipy.stats import spearmanr

#  Download S&P 500 daily data
df_sp500 = yf.download("^GSPC", start="1998-10-12", end="2025-09-30")

# Flatten multiindex columns if needed
if isinstance(df_sp500.columns, pd.MultiIndex):
    df_sp500.columns = [col[0] for col in df_sp500.columns]

#  Compute daily log returns
price_col = "Adj Close" if "Adj Close" in df_sp500.columns else "Close"
df_sp500["log_ret"] = np.log(df_sp500[price_col] / df_sp500[price_col].shift(1))
log_returns = df_sp500["log_ret"].dropna()

# 3Spearman rank correlation test for weak-form EMH (random walk)
lagged_returns = log_returns.shift(1).dropna()
current_returns = log_returns.iloc[1:]

corr, pval = spearmanr(current_returns, lagged_returns)

print(f"Spearman rank correlation: {corr:.4f}")
print(f"P-value: {pval:.4g}")
if pval < 0.05:
    print("❌ Reject H0: returns are not independent (weak-form EMH violated)")
else:
    print("✅ Fail to reject H0: returns are independent (supports weak-form EMH)")

```

```python
import numpy as np
import pandas as pd
import statsmodels.api as sm

# --- assumes you already have df_sp500 with 'log_ret' computed ---
# e.g., from the earlier cell that downloads ^GSPC and creates df_sp500['log_ret']

def make_ar_data(ret_series, p):
    r = ret_series.dropna().to_numpy()  # numpy array
    # build lag matrix
    X = np.column_stack([np.roll(r, k) for k in range(1, p+1)])
    y = r.copy()
    # drop first p rows to align
    X = X[p:, :]
    y = y[p:]
    # add constant
    X = sm.add_constant(X)
    return y, X

def fit_ols(y, X):
    model = sm.OLS(y, X).fit()
    beta = model.params      # (p+1,)
    se   = model.bse         # (p+1,)
    return model, beta, se

def pairs_bootstrap_coefs(y, X, B=2000, random_state=0):
    rng = np.random.default_rng(random_state)
    n = len(y)
    k = X.shape[1]
    coefs = np.empty((B, k))
    for b in range(B):
        idx = rng.integers(0, n, size=n)  # sample rows with replacement
        yb = y[idx]
        Xb = X[idx, :]
        # OLS via lstsq; add small ridge if singular
        beta_b, *_ = np.linalg.lstsq(Xb, yb, rcond=None)
        coefs[b, :] = beta_b
    return coefs  # shape (B, k)

def ar_bootstrap_summary(ret_series, p, B=2000):
    y, X = make_ar_data(ret_series, p)
    model, beta, se = fit_ols(y, X)
    coefs_b = pairs_bootstrap_coefs(y, X, B=B, random_state=42)

    # Position-based arrays (no Pandas alignment)
    beta_np = np.asarray(beta)           # (p+1,)
    se_np   = np.asarray(se)             # (p+1,)
    coefs_np = np.asarray(coefs_b)       # (B, p+1)

    # Percentile CIs for slope coefficients (exclude constant at index 0)
    pct_lo, pct_hi = np.percentile(coefs_np[:, 1:], [2.5, 97.5], axis=0)

    # Studentized-t using *fixed* SE from original sample (quick version)
    boot_t = (coefs_np[:, 1:] - beta_np[1:]) / se_np[1:]   # (B, p)
    t_lo, t_hi = np.percentile(boot_t, [2.5, 97.5], axis=0)
    ci_t_lo = beta_np[1:] - t_hi * se_np[1:]
    ci_t_hi = beta_np[1:] - t_lo * se_np[1:]

    # Summary DataFrame
    names = [f"lag{k}" for k in range(1, p+1)]
    out = pd.DataFrame({
        "beta_hat": beta_np[1:],
        "se_ols":  se_np[1:],
        "t_stat":  beta_np[1:] / se_np[1:],
        "CI_pct_2.5": pct_lo,
        "CI_pct_97.5": pct_hi,
        "CI_t_2.5": ci_t_lo,
        "CI_t_97.5": ci_t_hi
    }, index=names)
    return out, model

# Run for AR(1), AR(3), AR(9)
summ_ar1, model_ar1 = ar_bootstrap_summary(df_sp500["log_ret"], p=1, B=2000)
summ_ar3, model_ar3 = ar_bootstrap_summary(df_sp500["log_ret"], p=3, B=2000)
summ_ar9, model_ar9 = ar_bootstrap_summary(df_sp500["log_ret"], p=9, B=2000)

print("AR(1) summary:\n", summ_ar1, "\n")
print("AR(3) summary:\n", summ_ar3, "\n")
print("AR(9) summary:\n", summ_ar9.head(), " ...")  # too wide to print all

```

```python
# === AR(p) Bootstrap Toolkit: S&P 500 daily ===
# Requires: pandas, numpy, statsmodels, yfinance, scipy (only for Spearman if you add it)

import numpy as np
import pandas as pd
import statsmodels.api as sm
import yfinance as yf

# -----------------------
# 0) Data: S&P500 ^GSPC
# -----------------------
def load_sp500(start="1998-10-12", end="2025-09-30"):
    df = yf.download("^GSPC", start=start, end=end, progress=False)
    # Flatten MultiIndex columns if yfinance returns them
    if isinstance(df.columns, pd.MultiIndex):
        df.columns = [c[0] for c in df.columns]
    # Use Adj Close if present, else Close
    price_col = "Adj Close" if "Adj Close" in df.columns else "Close"
    df["log_ret"] = np.log(df[price_col] / df[price_col].shift(1))
    df = df.dropna(subset=["log_ret"])
    return df

# -----------------------
# 1) Build AR(p) design
# -----------------------
def make_ar_design(ret_series: pd.Series, p: int):
    """Return y, X for OLS of r_t on [const, r_{t-1},...,r_{t-p}]"""
    r = ret_series.dropna().to_numpy()  # T,
    # lag matrix T x p
    Xlags = np.column_stack([np.roll(r, k) for k in range(1, p+1)])
    y = r.copy()
    # align: drop first p rows
    X = Xlags[p:, :]
    y = y[p:]
    # add constant (first col)
    X = sm.add_constant(X)
    return y, X  # shapes: (T-p,), (T-p, p+1)

def fit_ols(y, X):
    model = sm.OLS(y, X).fit()
    beta = np.asarray(model.params)  # (p+1,)
    se   = np.asarray(model.bse)     # (p+1,)
    return model, beta, se

# -----------------------
# 2) Pairs bootstrap
# -----------------------
def pairs_bootstrap_coefs(y, X, B=2000, random_state=0):
    rng = np.random.default_rng(random_state)
    n, k = X.shape
    coefs = np.empty((B, k))
    for b in range(B):
        idx = rng.integers(0, n, size=n)  # sample rows with replacement
        Xb = X[idx, :]
        yb = y[idx]
        # OLS; if singular, lstsq gives least-squares solution
        beta_b, *_ = np.linalg.lstsq(Xb, yb, rcond=None)
        coefs[b, :] = beta_b
    return coefs  # (B, k)

# -----------------------
# 3) Circular moving-block bootstrap
# -----------------------
def _circular_block_indices(n, L, B, rng):
    nb = int(np.ceil(n / L))  # number of blocks to cover n obs
    starts = rng.integers(0, n, size=(B, nb))
    base = np.arange(L)
    idx = np.empty((B, nb * L), dtype=int)
    for b in range(B):
        blocks = (starts[b, :, None] + base[None, :]) % n
        idx[b] = blocks.reshape(-1)
    return idx[:, :n]  # (B, n)

def block_bootstrap_coefs(y, X, B=2000, L=18, random_state=42):
    rng = np.random.default_rng(random_state)
    n, k = X.shape
    idx_all = _circular_block_indices(n, L, B, rng)  # (B, n)
    coefs = np.empty((B, k))
    for b in range(B):
        ib = idx_all[b]
        Xb = X[ib, :]
        yb = y[ib]
        beta_b, *_ = np.linalg.lstsq(Xb, yb, rcond=None)
        coefs[b, :] = beta_b
    return coefs  # (B, k)

# -----------------------
# 4) CI construction
# -----------------------
def summarize_bootstrap(beta, se, coefs, p):
    """
    beta: (p+1,) full coef (const + p lags)
    se:   (p+1,) OLS standard errors
    coefs:(B, p+1) bootstrap coefs
    returns DF for lag coefficients only
    """
    # Percentile CIs (exclude constant col 0)
    pct_lo, pct_hi = np.percentile(coefs[:, 1:], [2.5, 97.5], axis=0)

    # Studentized-t (using original SEs; fast & standard)
    boot_t = (coefs[:, 1:] - beta[1:]) / se[1:]  # (B, p)
    t_lo, t_hi = np.percentile(boot_t, [2.5, 97.5], axis=0)
    ci_t_lo = beta[1:] - t_hi * se[1:]
    ci_t_hi = beta[1:] - t_lo * se[1:]

    names = [f"lag{k}" for k in range(1, p+1)]
    out = pd.DataFrame({
        "beta_hat": beta[1:],
        "se_ols":   se[1:],
        "t_stat":   beta[1:] / se[1:],
        "CI_pct_2.5":  pct_lo,
        "CI_pct_97.5": pct_hi,
        "CI_t_2.5":    ci_t_lo,
        "CI_t_97.5":   ci_t_hi
    }, index=names)
    return out

# -----------------------
# 5) One-call wrappers
# -----------------------
def ar_pairs_bootstrap(ret_series, p=3, B=2000, random_state=0, verbose=True):
    y, X = make_ar_design(ret_series, p)
    model, beta, se = fit_ols(y, X)
    coefs = pairs_bootstrap_coefs(y, X, B=B, random_state=random_state)
    out = summarize_bootstrap(beta, se, coefs, p)
    if verbose:
        print(f"[Pairs bootstrap] AR({p}) on T={len(ret_series.dropna())} returns")
        print(f"B={B}")
        print(out.to_string())
    return out, model, coefs

def ar_block_bootstrap(ret_series, p=3, B=2000, L=18, random_state=42, verbose=True):
    y, X = make_ar_design(ret_series, p)
    model, beta, se = fit_ols(y, X)
    coefs = block_bootstrap_coefs(y, X, B=B, L=L, random_state=random_state)
    out = summarize_bootstrap(beta, se, coefs, p)
    if verbose:
        print(f"[Block bootstrap] AR({p}) on T={len(ret_series.dropna())} returns")
        print(f"B={B}, L={L}")
        print(out.to_string())
    return out, model, coefs

# -----------------------
# 6) Run examples
# -----------------------
if __name__ == "__main__":
    df = load_sp500()
    print("S&P500 date range:", df.index.min().date(), "→", df.index.max().date())
    print("Columns:", df.columns.tolist(), "\n")

    # Quick smoke test: smaller B
    _ = ar_pairs_bootstrap(df["log_ret"], p=1, B=300)
    _ = ar_block_bootstrap(df["log_ret"], p=3, B=300, L=18)

    # Final runs (uncomment to do full, may take longer)
    # _ = ar_pairs_bootstrap(df["log_ret"], p=3, B=2000)
    # _ = ar_block_bootstrap(df["log_ret"], p=9, B=2000, L=18)
```

## Problem 4 – Predictability of the Equity Premium

As my professor instructed us in the problem set, i used the updated monthly dataset from Amit Goyal’s webpage (provided locally as `PredictorData2023.xlsx – Monthly.csv`).  Following Goyal and Welch (2008), I construct 13 predictors for the S&P 500 equity premium:

1. **d/p** – dividend‑price ratio: \(\log(D_{12}) - \log(P)\).
2. **d/y** – dividend‑yield: \(\log(D_{12}) - \log(P_{t-1})\).
3. **e/p** – earnings‑price ratio: \(\log(E_{12}) - \log(P)\).
4. **d/e** – dividend‑earnings ratio: \(\log(D_{12}) - \log(E_{12})\).
5. **svar** – stock variance (sum of squared daily returns).
6. **csp** – cross‑sectional beta premium.
7. **b/m** – book‑to‑market ratio.
8. **ntis** – net equity issuance.
9. **tbl** – treasury bill rate.
10. **ltr** – long‑term government bond return.
11. **lty** – long‑term government bond yield.
12. **tms** – term spread (lty − tbl).
13. **infl** – inflation.

The target variable is the one‑month‑ahead equity premium: \(r_{t+1} - r^f_t\), where \(r_{t+1}\) is the CRSP value‑weighted return including dividends and \(r^f_t\) is the one‑month risk‑free rate.

I first construct the predictors and compute the in‑sample predictive regressions using all available data (starting in **1926‑12** for most variables and **1937‑05** for `csp`).  Each regression takes the form:

\[ r_{t+1} - r^f_t = lpha + eta X_t +
arepsilon_{t+1}, \]

where \(X_t\) is one of the predictors.  The coefficients, t‑statistics and \(R^2\) values are reported below.

```python

import pandas as pd
import numpy as np
import statsmodels.api as sm

# Load the provided monthly data (CSV file should be in the same directory)
df = pd.read_csv('PredictorData2023.xlsx - Monthly.csv')

# Convert numeric columns
for col in df.columns:
    df[col] = pd.to_numeric(df[col], errors='coerce')

# Convert yyyymm to a proper date (last day of the month)
df['year'] = df['yyyymm'] // 100
df['month'] = df['yyyymm'] % 100
df['date'] = pd.to_datetime(dict(year=df['year'], month=df['month'], day=1)) + pd.offsets.MonthEnd(0)

# Convert percentages to decimals for returns, yields and rates
for col in ['CRSP_SPvw','CRSP_SPvwx','Rfree','tbl','AAA','BAA','lty','ltr','corpr','infl']:
    df[col] = df[col] / 100.0

# Construct predictors
df['dp']  = np.log(df['D12']) - np.log(df['Index'])
df['dy']  = np.log(df['D12']) - np.log(df['Index'].shift(1))
df['ep']  = np.log(df['E12']) - np.log(df['Index'])
df['de']  = np.log(df['D12']) - np.log(df['E12'])
df['tms'] = df['lty'] - df['tbl']
df['dfy'] = df['BAA'] - df['AAA']
df['dfr'] = df['corpr'] - df['ltr']
df['bm']  = df['b/m']

# Excess return and one‑step ahead target
df['excess_return'] = df['CRSP_SPvw'] - df['Rfree']
df['excess_return_next'] = df['excess_return'].shift(-1)

predictors = ['dp','dy','ep','de','svar','csp','bm','ntis','tbl','ltr','lty','tms','infl']

results = []

for pred in predictors:
    sub = df[['date', 'excess_return_next', pred]].dropna()
    # Align to predict r_{t+1}
    y = sub['excess_return_next']
    X = sm.add_constant(sub[pred])
    model = sm.OLS(y, X).fit()
    results.append({
        'Predictor': pred,
        'Beta': model.params[pred],
        't‑stat': model.tvalues[pred],
        'p‑value': model.pvalues[pred],
        'R2': model.rsquared,
        'Observations': int(model.nobs)
    })

results_df = pd.DataFrame(results)
results_df
```

### Out‑of‑sample (OOS) forecasting

I evaluate each predictive regression out‑of‑sample from **January 1962** through **December 2024**.  The benchmark model is the historical average (constant‑only regression).  For each predictor we compute one‑step‑ahead forecasts under three estimation schemes:

1. **Fixed:** estimate the model once using data up to 1961‑12; use the same coefficients for all forecasts.
2. **Recursive:** expand the estimation window by one month each forecast date (i.e., use data up to time \(t\) to forecast \(t+1\)).
3. **Rolling:** use a 10‑year (120‑month) rolling window to estimate the model for each forecast.

For each scheme we compute:

- the out‑of‑sample coefficient of determination \(R^2_{OS} = 1 -
rac{\mathrm{MSE}_{	ext{model}}}{\mathrm{MSE}_{	ext{benchmark}}}\),
- the **Diebold–Mariano (DM) t‑statistic** based on the difference of squared forecast errors, and
- the **MSE‑F statistic** from McCracken (2007).

The following cell implements these calculations.  The resulting table (`oos_results_df`) shows the number of OOS observations and the statistics for each predictor and scheme.

```python
import pandas as pd, numpy as np
import statsmodels.api as sm

# === Config ===
PATH = "PredictorData2023.xlsx - Monthly.csv"  # change if needed
PREDICTORS_REQ = ['dp','dy','ep','de','svar','csp','bm','ntis','tbl','ltr','lty','tms','infl']
START_EVAL = pd.Timestamp('1962-01-31')

def _norm(s):
    return (s.strip().lower()
            .replace(' ', '').replace('/', '')
            .replace('\\','').replace('-','').replace('.',''))

# --- Load & normalize
raw = pd.read_csv(PATH)
df = raw.rename(columns={c: _norm(c) for c in raw.columns}).copy()

# Date
if 'yyyymm' not in df.columns:
    raise ValueError("Could not find 'yyyymm' in the file.")
df['year']  = (df['yyyymm'] // 100).astype(int)
df['month'] = (df['yyyymm'] % 100).astype(int)
df['date']  = pd.to_datetime(dict(year=df['year'], month=df['month'], day=1)) + pd.offsets.MonthEnd(0)

# Market & RF (robust find) → to decimal if needed
def first_existing(names):
    for n in names:
        if n in df.columns:
            return n
    return None

mkt_col = first_existing(['crsp_spvw','crsp_spvwi','crspspvw','crsp_spvwe'])
rf_col  = first_existing(['rfree','rf','rfr'])
if mkt_col is None or rf_col is None:
    raise ValueError("Missing market return (CRSP_SPvw*) and/or Rfree columns after normalization.")

for c in [mkt_col, rf_col]:
    df[c] = pd.to_numeric(df[c], errors='coerce')

def to_decimal(s):
    med = s.dropna().abs().median()
    return s/100.0 if (med is not None and med > 1.0) else s

df['mkt_ret'] = to_decimal(df[mkt_col])
df['rf']      = to_decimal(df[rf_col])
df['excess_return']      = df['mkt_ret'] - df['rf']
df['excess_return_next'] = df['excess_return'].shift(-1)

# Build tms if available inputs exist
if ('lty' in df.columns) and ('tbl' in df.columns) and ('tms' not in df.columns):
    df['tms'] = pd.to_numeric(df['lty'], errors='coerce') - pd.to_numeric(df['tbl'], errors='coerce')

# Predictors actually present
present_preds = [p for p in PREDICTORS_REQ if p in df.columns]
missing_preds = [p for p in PREDICTORS_REQ if p not in df.columns]

print(f"Data range: {df['date'].min().date()} → {df['date'].max().date()}")
print("Predictors present:", present_preds)
if missing_preds:
    print("Missing predictors (skipped):", missing_preds)

# --- Recursive OOS with forced constant for single-row predict
def recursive_oos_r2_forceconst(df, pred, start_date, min_est=30):
    sub = df[['date', pred, 'excess_return_next']].dropna().reset_index(drop=True)
    idxs = sub.index[sub['date'] >= start_date]
    if len(idxs) == 0:
        return None, 0
    start_idx = int(idxs[0])

    yhat_model, yhat_bench, y_true = [], [], []
    xvals = sub[pred].to_numpy(dtype=float)
    yvals = sub['excess_return_next'].to_numpy(dtype=float)

    for t in range(start_idx, len(sub)):
        est_y = yvals[:t]      # up to t-1 inclusive
        est_x = xvals[:t]

        # Drop NaNs (already dropped in sub, but be safe)
        mask = ~np.isnan(est_y) & ~np.isnan(est_x)
        est_y = est_y[mask]
        est_x = est_x[mask]
        if est_y.size < min_est:
            continue

        # Build X with forced constant for estimation
        X_est = sm.add_constant(est_x.reshape(-1,1), has_constant='add')
        model = sm.OLS(est_y, X_est).fit()

        # --- FORCE constant for the 1-row prediction too ---
        x_t = np.array([[1.0, xvals[t]]], dtype=float)  # shape (1,2)
        yhat_model.append(float(model.predict(x_t)))

        # Historical mean benchmark
        yhat_bench.append(float(est_y.mean()))

        # Actual
        y_true.append(yvals[t])

    n = min(len(y_true), len(yhat_model), len(yhat_bench))
    if n == 0:
        return None, 0

    y_true  = np.asarray(y_true[:n])
    y_model = np.asarray(yhat_model[:n])
    y_bench = np.asarray(yhat_bench[:n])

    mse_m = np.mean((y_true - y_model)**2)
    mse_b = np.mean((y_true - y_bench)**2)
    r2_os = 1.0 - mse_m/mse_b
    return r2_os, n

# --- Run and print results
results = []
for pred in present_preds:
    r2_os, n = recursive_oos_r2_forceconst(df, pred, START_EVAL)
    if n == 0 or r2_os is None or np.isnan(r2_os):
        print(f"[{pred}] no valid OOS window or insufficient data; skipped.")
        continue
    print(f"[{pred}] R2_os={r2_os:.4f}  (n_oos={n})")
    results.append({'predictor': pred, 'n_oos': n, 'R2_os': r2_os})

res_df = pd.DataFrame(results).sort_values('R2_os', ascending=False).reset_index(drop=True)
if res_df.empty:
    print("\nNo OOS results produced. Try a later START_EVAL (e.g., '1970-01-31') or check missing predictors.")
else:
    print("\n=== Out-of-sample R2 vs historical mean (recursive, forced-const) ===")
    print(res_df.to_string(index=False))

```

### White’s (2000) Reality Check

Because multiple predictive models are considered simultaneously, it is important to adjust for data‑snooping when assessing which model performs best.  White’s (2000) reality check uses a bootstrap procedure to test the null hypothesis that none of the candidate models outperforms the benchmark.

A simple implementation is as follows:

1. Compute the vector of forecast error losses for each model relative to the benchmark (here the squared error difference: \(d_{i,t} = e_{	ext{benchmark},t}^2 - e_{i,t}^2\)).
2. For each bootstrap replication, resample the sequence of loss differentials (e.g., using a moving block bootstrap to account for autocorrelation) and compute the maximal average differential across models.
